Q2. Write a python program to print Python Exception Hierarchy.

```python
import inspect
import builtins

def print_exception_hierarchy(base_class, indent=0):
    # Print the class name with indentation
    print(" " * indent + base_class.__name__)
    
    # Iterate over the subclasses of the base class
    for subclass in base_class.__subclasses__():
        print_exception_hierarchy(subclass, indent + 4)

def main():
    # Print the hierarchy starting from the base Exception class
    print("Python Exception Hierarchy:")
    print_exception_hierarchy(Exception)

if __name__ == "__main__":
    main()
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

```python
#1. KeyError
my_dict = {'name': 'Alice', 'age': 30}

try:
    value = my_dict['address']  # Key 'address' does not exist in the dictionary
except KeyError as e:
    print(f"KeyError: {e}")
```

```python
#2. IndexError
my_list = [1, 2, 3]

try:
    value = my_list[5]  # Index 5 is out of range for the list
except IndexError as e:
    print(f"IndexError: {e}")
```
